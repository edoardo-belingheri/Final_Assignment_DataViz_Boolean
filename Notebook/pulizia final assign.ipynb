```python
climate['elgnuc'] = climate['elgnuc'].astype('Int32')
climate['elgsun'] = climate['elgsun'].astype('Int32')
climate['elgwind'] = climate['elgwind'].astype('Int32')
climate['elgbio'] = climate['elgbio'].astype('Int32')
```

```python
climate.rename(columns={'elgnuc': 'quota_elettr_da_nucleare',
                        'elgsun': 'quota_elettr_da_solare',
                        'elgwind': 'quota_elettr_da_eolico',
                        'elgbio': 'quota_elettr_da_biomassa',
                        }, inplace=True)
```

```python
lista_quote=['quota_elettr_da_carbone', 'quota_elettr_da_gas_naturale', 'quota_elettr_da_idroelettrico', 'quota_elettr_da_nucleare', 'quota_elettr_da_solare',
            'quota_elettr_da_eolico', 'quota_elettr_da_biomassa']

def riclassifica_quantit(val):
    if val in [1, 2]:
        return "Large"
    elif val == 3:
        return "Medium"
    elif val in [4, 5]:
        return "Small"
    else:
        return pd.NA  # mantiene i NaN e i codici fuori scala

for el in lista_quote:
    climate[el] = climate[el].apply(riclassifica_quantit)
```
